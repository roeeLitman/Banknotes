```python
!pip install -qU langchain langgraph langchain-groq
import os
from google.colab import userdata
from langchain_groq import ChatGroq
from langgraph.graph import StateGraph, END, START
from langchain_core.prompts import ChatPromptTemplate
from pydantic import BaseModel, Field
from enum import Enum

os.environ["GROQ_KEY"] = userdata.get('GROQ_KEY')
```

```python
class Route(str, Enum):
    HUMAN_REVIEW = "human_review"
    FINISH = "finish"

class Category(str, Enum):
    SUPPORT = "Support"
    SALES = "Sales"
    SPAM = "Spam"
    OTHER = "Other"

class EmailAnalysis(BaseModel):
    classification: Category = Field(
        default=Category.OTHER,
        description="The primary category. Use SUPPORT, SALES, or SPAM. If it doesn't fit, use OTHER."
    )
    reasoning: str = Field(
        default=None,
        description="Briefly explain your reasoning for this classification."
    )
    suggest_new_category: bool = Field(
        default=False,
        description="Set to TRUE if classification is OTHER or if you believe a new category is needed for this topic."
    )
    new_category_name: str = Field(
        default="",
        description="If suggest_new_category is True, provide a concise name for the new category (e.g., 'SpaceLegal'). Otherwise, leave empty."
    )

class State(EmailAnalysis):
    email_text: str



llm = ChatGroq(
    temperature = 0,
    model_name = "openai/gpt-oss-120b",
    api_key = os.environ.get("GROQ_KEY"),
)

structured_llm = llm.with_structured_output(EmailAnalysis)

system_prompt = """
You are a professional business email classifier.
Your goal is to categorize incoming mail into [SUPPORT, SALES, SPAM, OTHER].

### 1. SALES (High Priority)
Classify as SALES if the sender shows clear 'Buying Intent'. This includes:
- Pricing inquiries, quotes, or requests for bulk discounts.
- Interest in purchasing licenses, products, or new services.
- Requests for a demo, a sales call, or partnership proposals.
- Questions about product availability or "How do I buy...".
*Note: If the sales inquiry is for a service we DON'T currently offer (e.g., legal advice for cats), classify as OTHER and set suggest_new_category to True.*

### 2. SUPPORT
Classify as SUPPORT if the sender has a technical problem, needs help with an existing account, or reports a bug/issue with a product they already have.

### 3. SPAM
Classify as SPAM only for unsolicited marketing, phishing, or generic junk. If someone is genuinely trying to contact the business (even if it's a weird request), do NOT call it Spam.

### 4. OTHER & HUMAN REVIEW PROTOCOL
- Use OTHER for legitimate business messages that don't fit Sales or Support.
- Whenever you pick OTHER, you MUST set 'suggest_new_category' to True.
- Provide a potential name for this new category in 'new_category_name'.

DO NOT be "helpful" by forcing a weird Sales lead into the Sales category. If it's outside our domain, use OTHER + suggest_new_category=True.
"""

prompt_template = ChatPromptTemplate.from_messages([
    ("system", system_prompt),
    ("user", "Email content: {email_text}")
])

def classifier_node(state: State):
    prompt = prompt_template.format_messages( email_text = state.email_text )
    response = structured_llm.invoke(prompt)

    return {
        "classification": response.classification,
        "reasoning": response.reasoning,
        "suggest_new_category": response.suggest_new_category,
        "new_category_name": response.new_category_name
    }

def human_review(state: State):
    print("-" * 10 + " human_review " + "-" * 10 + "\n")
    print(f"new_category_name {state.new_category_name}")
    return state

def is_create_new_classification(state: State):
    if state.suggest_new_category:
        return Route.HUMAN_REVIEW
    return Route.FINISH

workflow = StateGraph(State)

workflow.add_node("classifier", classifier_node)
workflow.add_node("human_review", human_review)

workflow.add_edge(START, "classifier")

workflow.add_conditional_edges(
    "classifier",
    is_create_new_classification,
    {
        Route.HUMAN_REVIEW: "human_review",
        Route.FINISH: END
    }
)
workflow.add_edge("human_review", END)

app = workflow.compile()

test_emails = [
    # TEST CASE 1: Standard Support
    # Purpose: Verify that routine technical issues are classified correctly and skip human review.
    "My internet is down and I've tried restarting the router. Help!",

    # TEST CASE 2: High-Intent Sales
    # Purpose: Check if the AI recognizes "Buying Intent" (licenses, bulk pricing) and routes to Sales.
    "I'm interested in purchasing 50 licenses for your software. Do you have bulk pricing?",

    # TEST CASE 3: Obvious Spam
    # Purpose: Ensure that generic junk mail is filtered out immediately without bothering a human.
    "CONGRATULATIONS! You won a million dollars! Click here to claim your prize now!!!",

    # TEST CASE 4: Out-of-Distribution (The "Space Cat")
    # Purpose: This is the "Edge Case". It should trigger 'suggest_new_category=True' and PAUSE for human review.
    "I want to know if your company provides legal advice for space-traveling cats. Is there a lawyer for Mars?",

    # TEST CASE 5: Robustness & Prank
    # Purpose: Test if the AI can handle contradictory intent (Support + Insult) and decide if it's Spam or Other.
    "I need help with my broken laptop... just kidding, you are a stupid robot!"
]

print("--- STARTING SYSTEM BATCH TEST ---\n")

for email in test_emails:
    print(f"📧 ANALYZING EMAIL: '{email[:50]}...'")

    result = app.invoke({"email_text": email})

    print(f"✅ FINAL DECISION: {result['classification']}")
    print(f"📝 REASONING: {result['reasoning']}")
    print(f"🔎 SUGGEST NEW CATEGORY: {result['suggest_new_category']}")
    print("-" * 50)

print("\n--- ✅ BATCH TEST COMPLETED ---")
```
